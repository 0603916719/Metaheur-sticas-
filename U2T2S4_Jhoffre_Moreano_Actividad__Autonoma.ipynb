#1. Selección del problema benchmark

Seleccioné el problema Traveling Salesman Problem (TSP) en una versión reducida de 6 nodos. El objetivo es encontrar la ruta más corta que permita visitar todas las ciudades una sola vez y regresar al punto de inicio.

# 2. Implementación de algoritmos de trayectoria
Se implementaron dos algoritmos:
•	Simulated Annealing (SA)
•	Tabu Search (TS).


```python
#Jhoffre Moreano

import random
import math

# ==============================
# DATOS DEL PROBLEMA TSP
# ==============================

# Matriz de distancias entre 6 ciudades
distancias = [
    [0, 10, 15, 20, 25, 30],
    [10, 0, 35, 25, 17, 28],
    [15, 35, 0, 30, 20, 26],
    [20, 25, 30, 0, 18, 24],
    [25, 17, 20, 18, 0, 16],
    [30, 28, 26, 24, 16, 0]
]

# ==============================
# FUNCIÓN DE COSTO
# ==============================

def calcular_costo(ruta):
    costo = 0

    for i in range(len(ruta) - 1):
        costo += distancias[ruta[i]][ruta[i + 1]]

    costo += distancias[ruta[-1]][ruta[0]]

    return costo


# ==============================
# GENERAR VECINO
# ==============================

def generar_vecino(ruta):
    nueva_ruta = ruta.copy()

    i, j = random.sample(range(len(ruta)), 2)

    nueva_ruta[i], nueva_ruta[j] = nueva_ruta[j], nueva_ruta[i]

    return nueva_ruta


# ==============================
# ALGORITMO 1: SIMULATED ANNEALING
# ==============================

def simulated_annealing():
    ruta_actual = list(range(len(distancias)))
    random.shuffle(ruta_actual)

    mejor_ruta = ruta_actual.copy()
    mejor_costo = calcular_costo(mejor_ruta)

    temperatura = 1000
    enfriamiento = 0.95
    iteraciones = 100

    while temperatura > 1:
        for _ in range(iteraciones):
            vecino = generar_vecino(ruta_actual)

            costo_actual = calcular_costo(ruta_actual)
            costo_vecino = calcular_costo(vecino)

            diferencia = costo_vecino - costo_actual

            if diferencia < 0 or random.random() < math.exp(-diferencia / temperatura):
                ruta_actual = vecino

            if calcular_costo(ruta_actual) < mejor_costo:
                mejor_ruta = ruta_actual.copy()
                mejor_costo = calcular_costo(mejor_ruta)

        temperatura *= enfriamiento

    return mejor_ruta, mejor_costo


# ==============================
# ALGORITMO 2: TABU SEARCH
# ==============================

def tabu_search():
    ruta_actual = list(range(len(distancias)))
    random.shuffle(ruta_actual)

    mejor_ruta = ruta_actual.copy()
    mejor_costo = calcular_costo(mejor_ruta)

    lista_tabu = []
    tamaño_tabu = 5
    iteraciones = 200

    for _ in range(iteraciones):
        vecinos = []

        for _ in range(20):
            vecino = generar_vecino(ruta_actual)

            if vecino not in lista_tabu:
                vecinos.append(vecino)

        if len(vecinos) == 0:
            continue

        mejor_vecino = min(vecinos, key=calcular_costo)
        ruta_actual = mejor_vecino

        costo_actual = calcular_costo(ruta_actual)

        if costo_actual < mejor_costo:
            mejor_ruta = ruta_actual.copy()
            mejor_costo = costo_actual

        lista_tabu.append(ruta_actual)

        if len(lista_tabu) > tamaño_tabu:
            lista_tabu.pop(0)

    return mejor_ruta, mejor_costo


# ==============================
# EJECUCIÓN DE LOS ALGORITMOS
# ==============================

ruta_sa, costo_sa = simulated_annealing()
ruta_ts, costo_ts = tabu_search()

print("Resultado con Simulated Annealing")
print("Mejor ruta:", ruta_sa)
print("Costo total:", costo_sa)

print("\nResultado con Tabu Search")
print("Mejor ruta:", ruta_ts)
print("Costo total:", costo_ts)
```
